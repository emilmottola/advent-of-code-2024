```python
# Test Data
# file_path = "day2_test.txt"

# More Test Data
# file_path = "day2_test_2.txt"

# Real Data
file_path = "day2.txt"

reports = []

with open(file_path, "r") as file:
    for line in file:
        # Split each line into numbers and convert them to integers
        row = [int(value) for value in line.strip().split()]
        # Append the row to the list of rows
        reports.append(row)

# Print the resulting list of lists
len(reports)
```

# Part 1

```python
safe = 0
count = 0
for report in reports:
    count += 1
    print(report)
    rep_ok = 0
    info = report[1] - report[0]
    # print(f"info is {info}")
    if info >= 1:
        direction = "up"
    else:
        direction = "down"
    # print(f"rep_ok is {rep_ok}")
    # print(f"directon is {directon}")    
    for i in range(len(report)-1):
        diff = report[i+1] - report[i]
        # print(f'diff is {diff}')
        if diff > 3 or diff < -3 or diff == 0:
            rep_ok = 0
            print("Unacceptable Change")
            break
        if direction == "up" and diff < 0:
            rep_ok = 0
            print("Wrong Direction")
            break
        if direction == "down" and diff > 0:
            rep_ok = 0
            print("Wrong Direction")
            break
        rep_ok = 1
    if rep_ok:
        print("Totally safe")
        safe += 1
```

# Part 2

```python
def check_report(report, instant_death=0):
    print(report)
    # print(f"info is {info}")
    direction = "up" if report[1] > report[0] else "down"
    # print(f"direction is {direction}")
    bad_step = 0
    for i in range(len(report)-1):
        diff = report[i+1] - report[i]
        # print(f'diff is {diff}')
        if diff > 3 or diff < -3:
            bad_step += 1
            print("Step too large")
        if  diff == 0:
            bad_step += 1
            print("Same Value")
        if direction == "up" and diff < 0:
            bad_step += 1
            print("Direction Changed to Down")
        if direction == "down" and diff > 0:
            bad_step += 1
            print("Direction Changed to Up")
        if bad_step > 0:
            if instant_death:
                return 0
            else:
                report_0 = report[1:]
                check_0 = check_report(report_0, instant_death=1)
                print(f"can we remove Element 0?: {check_0}")
                if check_0:
                    return 1
                report_a = report[:i] + report[i+1:]
                check_a = check_report(report_a, instant_death=1)
                print(f"can we remove A?: {check_a}")
                if check_a:
                    return 1
                report_b = report[:i+1] + report[i+2:]
                check_b = check_report(report_b, instant_death=1)
                print(f"can we remove B?: {check_b}")
                if check_b:
                    return 1
                else: 
                    return 0
    return 1
```

```python
safe = 0
count = 0
for report in reports:
    count += 1
    # print(report)
    # print(f"rep_ok is {rep_ok}")
    # print(f"directon is {directon}")    
    rep_ok = check_report(report)
    if rep_ok:
        print("Totally safe")
        safe += 1
        print(safe)
    else:
        print("UNSAFE AF")
```
